1) Crear una función que reciba un número como parámetro y devuelva True si es primo y False si no lo es

```python
def primo(numero):
    for a in range(2,(int(numero/2)+1)):
        if numero%a==0:
            return False
    else:
        return True
primo(13)
```

2) Utilizando la función del punto 1, realizar otra función que reciba de parámetro una lista de números y devuelva sólo aquellos que son primos en otra lista

```python
def identificar_primos(lista_numeros=[]):
    lista_de_primos=[]
    if lista_numeros == []:
        print("la lista esta vacía")
    for numero in lista_numeros:
        if primo(numero):
            lista_de_primos.append(numero)
    return lista_de_primos       

identificar_primos([2,3,4,5,6,7,8,9])
```

3) Crear una función que al recibir una lista de números, devuelva el que más se repite y cuántas veces lo hace. Si hay más de un "más repetido", que devuelva cualquiera

```python

def repeticion_numeros(lis_num_random):
    lista_auxiliar=set(lis_num_random)
    maximo=0
    numero_valor=0
    n_repeticiones=0
    for num in lista_auxiliar:
        n_repeticiones=0
        for i in lis_num_random:
            if i==num:
                n_repeticiones+=1
                if n_repeticiones>maximo:
                    maximo=n_repeticiones
                    numero_valor=num
    return numero_valor,maximo
repeticion_numeros([1,2,2,2,3,3,3,4,4,4,4,4,4,5])

    
```
