16) Crear una tupla que contenga los números enteros del 1 al 20

```python
#podria escribir esto tupla = (1,2,3,4,5,) pero lo hice con iterador por si se quiere usar otros valores en un futuro
listaAux=[]
cantidad= 21 #se podría agregar un input y pedir la cantidad de números
for a in range(1,cantidad): 
    listaAux.append(a)
primera_tupla=tuple(listaAux)    
print(primera_tupla)
```

18) Evaluar si los números 20 y 30 están dentro de la tupla

```python
cantidad = 0
for a in range(0,len(primera_tupla)):
    if(primera_tupla[a]==20 or primera_tupla[a]==30):
        cantidad+=1
if cantidad==0:
    print("no esta ninguno de los dos")
elif cantidad==1:
    print("solo esta 1")
else:
    print("no hay ninguno")

#como esta el 20 solo coincidirá con 1 
```
